##JOIN

```python
leftp = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
'B': ['B0', 'B1', 'B2', 'B3'],
'C': ['C0', 'C1', 'C2', 'C3'],
'D': ['D0', 'D1', 'D2', 'D3']},
index=[0, 1, 2, 3])
rightp = pd.DataFrame({'A': ['A0', 'A1', 'A6', 'A7'],
'F': ['B4', 'B5', 'B6', 'B7'],
'G': ['C4', 'C5', 'C6', 'C7'],
'H': ['D4', 'D5', 'D6', 'D7']},
index=[4, 5, 6, 7])
```

```python
lefts = spark.createDataFrame(leftp)
rights = spark.createDataFrame(rightp)
```

#### Lef Join

```python
leftp.merge(rightp, on ="A", how="left")
```

```python
lefts.join(rights, on = "a", how="left").show()
```

#### Right Join



```python
leftp.merge(rightp, on="A", how="right")
leftp.merge(rightp, on="A", how="inner")
leftp.merge(rightp, on="A", how="outer") # full


```

###  Concat Columns


```python
my_list = [('a', 2, 3),
('b', 5, 6),
('c', 8, 9),
('a', 2, 3),
('b', 5, 6),
('c', 8, 9)]
col_name = ['col1', 'col2', 'col3']
```

```python
dp1 = pd.DataFrame(my_list, columns = col_name)
dp1
```

```python
ds1 = spark.createDataFrame(my_list, schema = col_name)
ds1.show()
```

```python
dp1['concat'] = dp1.apply(lambda x: '%s%s'%(x['col1'],x['col2']), axis=1 )
dp1
```

```python
ds1.withColumn('concat', F.concat('col1', 'col2')).show()
```

### Window

```python
d = {'A':['a','b','c','d'],'B':['m','m','n','n'],'C':[1,2,3,6]}
dp = pd.DataFrame(d)
ds = spark.createDataFrame(dp)
```

```python
dp['rank'] = dp.groupby('B')['C'].rank('dense',ascending=False)
dp
```

```python
from pyspark.sql.window import Window
```

```python
w = Window.partitionBy('B').orderBy(ds.C.desc() )
```

```python
ds.withColumn('rank', F.rank().over(w) ).show()
```

### Rank vs Denserank


```python
d ={'Id':[1,2,3,4,5,6],
'Score': [4.00, 4.00, 3.85, 3.65, 3.65, 3.50]}
#
data = pd.DataFrame(d)
dp = data.copy()
ds = spark.createDataFrame(data)
```

```python
dp['dense_rank'] = dp['Score'].rank(method='dense', ascending = False)

dp['rank'] = dp["Score"].rank(method = 'min' , ascending= False) 
dp
```

```python
import pyspark.sql.functions as F
from pyspark.sql.window import Window
```

```python
w = Window.orderBy(ds.Score.desc())
```

```python
ds.withColumn('dense_rank', F.dense_rank().over(w) ).show()
```

```python
ds.withColumn('rank', F.rank().over(w) ).show()
```
