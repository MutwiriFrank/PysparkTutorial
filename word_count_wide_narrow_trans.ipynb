```python
from pyspark import SparkContext, SQLContext, SparkConf, StorageLevel
import numpy
```

```python
SparkContext.getOrCreate().stop()
sparkconf = SparkConf().setAppName("003").setMaster("local[3]") 
sc = SparkContext(conf= sparkconf)
```

```python
text = sc.textFile("/home/mutwiri/hadoop/spark/data/license.txt")
```

```python

# This job will be divided to two stages. This is because reduce by key by is a wide transformation and therefore the driver 
# has to wait for the first stage(flatmap and map) to be completed by all parirtitions before data is grouped.
# A new stage is created after every wide transformation

flat_text = text.flatMap(lambda line: line.split(" ")).map(lambda word: (word, 1)).reduceByKey( lambda a,b: a+b)

flat_text.collect()
```

```python
sc.stop()
```
